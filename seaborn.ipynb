### **Graficas Seaborn**

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# primero se importan las bibliotecas necesarias para realizar las graficas
```

```python
# Grafico lineplot
# se crea el dataframe con datos inventados y luego se ejecuta el grafico
df1 = pd.DataFrame({"size": [1, 2, 3, 4, 5, 6], "tip": [2, 3, 5, 4, 6, 7]})
sns.lineplot(data=df1, x="size", y="tip", color="purple")
plt.show()
```

```python
# Diagrama de cajas y bigotes
# Datos
df2 = pd.DataFrame({"day": ["Mon", "Mon", "Mon", "Tue", "Tue", "Tue","Wed", "Wed", "Wed", "Thu", "Thu", "Thu"], "total_bill": [9,15,18,8,17,22,16,19,22,18,15,25]})

# Boxplot
sns.boxplot(data=df2, x="day", y="total_bill", color="green")

plt.show()
```

```python
# Diagrama de violines
sns.violinplot(data=df2, x="day", y="total_bill", color="pink", inner="box")
plt.show()
```

```python
# Pairplot
sns.pairplot(df2)
plt.show()
```

```python
# Heatmap
# Se crea el df y luego se ejecuta el grafico
df = pd.DataFrame({"edad": [18, 20, 22, 24, 26, 28], "horas_estudio": [2, 3, 4, 5, 6, 7], "nota": [3.0, 3.5, 4.0, 4.2, 4.5, 4.8], "horas_sueo": [8, 7, 7, 6, 6, 5]})

corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="Oranges")
plt.show()
```
