```python
# Matplotlib

import matplotlib.pyplot as plt
import numpy as np

# 1. Grafico linea basico

x = [0, 2, 4, 6, 8]
y = [5, 15, 18, 22, 28]

plt.plot(x, y)
plt.title("Grafico lineal")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

```python
x1 = np.linspace(-3, 7, 100)
y1 = -1.5*x1 + 4

plt.plot(x1, y1)

plt.text(1, -6, r"$y = -1.5x + 4$", fontsize=12)
plt.text(1, -7.5, r"$R^2 = 0.95$", fontsize=12)

plt.title("Ecuaci√≥n conocida")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
x2 = np.array([2, 4, 6, 8, 10])
y2 = np.array([3.9, 8.2, 11.8, 16.1, 19.9])

# regresion lineal

m, b = np.polyfit(x2, y2, 1)
y2_pred = m*x2 + b

# calcular R^2

SS_res = np.sum((y2 - y2_pred)**2)
SS_tot = np.sum((y2 - np.mean(y2))**2)
r2 = 1 - ((SS_res / SS_tot))

plt.scatter(x2, y2)
plt.plot(x2, y2_pred, color='red')

eq = rf"$y = {m:.2f}x + {b:.2f}$"
r2_text = rf"$R^2 = {r2:.4f}$"

plt.text(2, 20, eq, fontsize=12)
plt.text(2, 17, r2_text, fontsize=12)

plt.title("Ecuacion conocida")
plt.xlabel("tiempo")
plt.ylabel("concentraci√≥n")
plt.grid(True)
plt.show()
```
